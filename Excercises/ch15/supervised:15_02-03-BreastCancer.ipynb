## Import Dataset

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
breast_cancer = load_breast_cancer()
print(breast_cancer.DESCR)
```

## Basic informations

```python
# Features of the dataset
breast_cancer.feature_names
```

```python
# Predictors of the dataset
breast_cancer.target_names
```

## Visualization

```python
# Labels of the dataset with data and the features
cancer_df = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
# Add the target column
cancer_df['target'] = breast_cancer.target
# Show the dataset
cancer_df.head()
```

```python
# Observations of the dataset: no null-values and all numeric values
cancer_df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sn
sn.heatmap(cancer_df.corr(), annot=False)
```

## Splitting the dataset

```python
from sklearn.model_selection import train_test_split
X = breast_cancer.data      # Predictors
y = breast_cancer.target    # Labels

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```
